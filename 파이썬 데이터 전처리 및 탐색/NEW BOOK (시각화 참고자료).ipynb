```python
data = pd.read_csv("all_include.csv", index_col = "date", parse_dates = True)
```

```python
data.index
```

### 2.1 위의 데이터를 활용하여 다음 그림과 같이 그래프를 그리세요
 - 표시된 COVID-19의 날짜는 "kopsi200"이 2018년 이후 최저가 된 시점.
 - color와 size는 개인들의 취향따라 정하세요
 

```python
min_kospi = data.loc["2018-01-01":,:]["kospi200"].min()
```

```python
min_date = data[data["kospi200"] == min_kospi].index
```

```python
data["kospi200"]
```

```python
from datetime import datetime
```

```python
a = datetime(2018,3,1)
a
```

```python
plt.figure(figsize = (15,5))
plt.plot(data.index, data["kospi200"])
plt.axhline(min_kospi, color = "red", linestyle = ":")
plt.annotate("COVID-19",xy = (min_date,min_kospi), xytext = (a,350),
            arrowprops = {"arrowstyle":"->", "connectionstyle":"angle3"},fontsize = 20, color = "purple")

plt.xlabel("date")
plt.ylabel("index")
plt.suptitle("KOSPI200", fontsize = 20)
```

### 2.2 위의 그래프에서 2018년1월1일부터 2022년10월1일 사이의 그래프만 출력하세요

```python
first = datetime(2018,1,1)
last = datetime(2022,10,1)
```

```python
plt.figure(figsize = (15,5))
plt.plot(data.index, data["kospi200"])
plt.axhline(min_kospi, color = "red", linestyle = ":")
plt.annotate("COVID-19",xy = (min_date,min_kospi), xytext = (a,350),
            arrowprops = {"arrowstyle":"->", "connectionstyle":"angle3"},fontsize = 20, color = "purple")

plt.xlabel("date")
plt.ylabel("index")
plt.suptitle("KOSPI200", fontsize = 20)

plt.xlim(first, last)
```

### 2.3 다음 그래프와 같이 그리시요
 - 데이터는 "kospi200"과 "NSI"라는 열을 사용한다.
 - x축을 공유한다


```python
fig, ax1 = plt.subplots(1,1, figsize = (20,6))
plt.suptitle("Visualizing", fontsize = 30)

ax1.plot(data["kospi200"], label = "KOSPI")
ax1.tick_params(axis = "y",labelcolor = "tab:blue")
ax1.set_ylabel("kospi200", color = "tab:blue", fontsize = 20)
ax1.legend(loc = (0.935,0.9))

ax2 = ax1.twinx()
ax2.plot(data["NSI"], color = "tab:red",label = "NSI")
ax2.tick_params(axis = "y",labelcolor = "tab:red")
ax2.set_ylabel("NSI", color = "tab:red", fontsize = 20)
ax2.legend(loc = (0.935,0.85))
```
