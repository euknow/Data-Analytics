```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2, include_bias = False)
poly_x = poly.fit_transform(x)

poly_x.shape
```

```python
model2 = LinearRegression()
fitted_model2 = model2.fit(poly_x,y)

y_pred2 = fitted_model2.predict(poly_x)

plt.scatter(x, y, s=5)
plt.scatter(x,y_pred2,color = "darkorange")

from sklearn.metrics import r2_score

print(f"결정계수 : {r2_score(y, y_pred2)}")
```

```python
error = y-y_pred2

plt.figure(figsize = (8,4))
plt.scatter(x, error.ravel(), alpha = 0.5)
plt.axhline(0, color = "tab:red", ls = "--", alpha = 0.7)
plt.ylim(-11,11)
```
